```python
import pandas as pd
import time 
from datetime import datetime
import random
import urllib
import sqlalchemy
import matplotlib.pyplot as plt
import sys

position = {}
remaining_cash = 100000.00
loss = 0
win = 0
total_valuation = 0
profit = 0
```

```python
params = urllib.parse.quote("DRIVER={ODBC Driver 13 for SQL Server};SERVER=DESKTOP-LSM7IMP\SQLEXPRESS;DATABASE=StocksDMW;Trusted_Connection=yes")
engine = sqlalchemy.create_engine('mssql+pyodbc:///?odbc_connect=%s' % params)
```

```python
sql1= """
select distinct(date) from stockdatadaily 
"""
dates = pd.read_sql(sql1,engine)['date'].tolist()
```

```python
class equity():
    bought = None
    support = None
    last_support_date = None
    resistance = None
    resistance_date = None
    stop_order = 0.0
    times_bought = 0
    symbol = None
    
    def __init__(self, name):
        self.symbol = name

sql1= """
select distinct(name) from stockdatadaily 
"""
names = pd.read_sql(sql1,engine)['name'].tolist()

stock_equities = list()
for name in names:
    stock_equities.append(equity(name))
print(names)
```

```python
def automatic_trading():
    global position
    global remaining_cash 
    global loss 
    global win
    global total_valuation
    global win
    global profit
    
    count = 0
    for date in dates:
        if count == 0:
            count = count + 1
            continue
                        
        
       
        sql1= """
        select distinct(name) from stockdatadaily where date = '%s' 
        """ % (date)
        
        symbols = pd.read_sql(sql1,engine)['name'].tolist()
        
        for symbol in symbols:
            start_time = time.time()
            current_symbol = equity_getter(symbol)
            if current_symbol.bought == None:
                if resist_condition(symbol, date):
                    current_symbol.resistance, current_symbol.resitance_date = resistance_cal(symbol, date)
                
                if current_symbol.resistance != None and support_condition(symbol, date, current_symbol.resistance):
                    current_symbol.support = support_cal(symbol, date, current_symbol.resistance_date)
                    current_symbol.stop_order = current_symbol.support * 0.97
                    
                    current_price_query = """
                    select [4. close] from stockdatadaily where name = '%s' and date = '%s'
                    
                    """ % (symbol, date)
                    print("Query")
                    current_price = float(pd.read_sql(current_price_query, engine).values[0])
                    current_symbol.bought = current_price
                    remaining_cash = remaining_cash - 0.02 * remaining_cash
                    quantity = (0.02 * remaining_cash) // current_price
                    position.update({symbol: quantity})
                    current_symbol.times_bought += 1
                    
                    print("Buying ---> ", symbol, "on Date::", date)
                    print("Bought at:: ", current_price)
            
            else:
                current_price_query = """
                select [4. close] from stockdatadaily where name = '%s' and date = '%s'
                    
                """ % (symbol, date)
                print("Query")
                current_price = float(pd.read_sql(current_price_query, engine).values[0]) 
                
                if current_symbol.stop_order > current_price:
                    current_symbol.stop_order = None
                    print("Selling ---> ", symbol, "On Date:", date)
                    Print("Sold at:: ", current_price)
                    
                    del position[symbol]
                    current_symbol.bought = None
                    
                else:
                    if resist_condition(symbol,date):
                        current_symbol.resistance, current_symbol.resitance_date = resistance_cal(symbol, date)
                    if current_symbol.resistance != None and support_condition(symbol, date, current_symbol.resistance):
                        current_symbol.support = support_cal(symbol, date, current_symbol.resistance_date)
                        current_symbol.stop_order = current_symbol.support * 0.97
            
            print("Symbol Time ::: -------> ", symbol, time.time() - start_time) 
```

```python
def equity_getter(name):
    for stock in stock_equities:
        if stock.symbol == name:
            return stock


def resistance_cal( symbol, date):
    start_time = time.time()
    resistance_query= """
    select max([2. high]) from StockDataDaily
    where name = '%s' and date <= '%s'
    """ % (symbol, date)
     
    resistance = float(pd.read_sql(resistance_query,engine).values[0])
    
    date_query = """
    select date from StockDataDaily where name = '%s' and [2. high] = %f
    """ % (symbol, resistance)
     
    resistance_date = pd.read_sql(date_query, engine)['date'].values[0]
    print("resistance Cal", symbol, time.time() - start_time)
    return resistance, resistance_date
  

def support_cal(symbol, today_date, resistance_date):
    start_time = time.time()
    support_query = """
    select min([3. low]) from stockdatadaily where name = '%s' and date > '%s'and date < '%s'
    """ % (symbol, resistance_date, today_date)
     
    support = pd.read_sql(support_query, engine).values[0]
    print("support Cal", symbol, time.time() - start_time)

    return float(support)

def resist_condition(symbol, date):
    start_time = time.time()
    resistance_CQuery= """
    select TOP 2 * from StockDataDaily where date <= '%s' and name = '%s' order by date desc
    """ % (date, symbol)
     
    recent_data = pd.read_sql(resistance_CQuery,engine)
    today_price = recent_data['3. low'].values[0]
    if(len(recent_data) > 1):
        yesterday_price = recent_data['3. low'].values[1]

    print("resistance Condition", symbol, time.time() - start_time)
    if(today_price < yesterday_price):
        return True
    else:
        return False
    

    # If today's lowest price of the stock is less than yesterday's lowest price of the stock,
    # then Resistance condition is met

def support_condition(symbol, date, resistance):
    start_time = time.time()
    support_query = """
    select [4. close] from stockdatadaily where name = '%s' and date = '%s' 
    """ % (symbol, date)
     
    today_price = pd.read_sql(support_query, engine)['4. close'].values[0]
    
        
    print("support Condition", symbol, time.time() - start_time)
    if today_price > resistance:
        return True
    else:
        return False
    

def buy_condition(self, resist, data):
    if resist < data.tail(1)['2. high'].values[0]:
        return True
    else:
        return False       
```
